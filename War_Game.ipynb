```python
import random
```

```python
from IPython.display import clear_output
```

```python
### Creating Card Class
#### Suit , Rank and Value

suits = ('Hearts' , 'Diamonds' , 'Spades' , 'Clubs')

ranks = ('Two','Three','Four','Five','Six','Seven','Eight',
         'Nine','Ten','Jack','Queen','King','Ace')

values = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,
         'Nine':9,'Ten':10,'Jack':11,'Queen':12,'King':13,'Ace':14}
```

```python
class Card():
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]

    def __str__(self):
        return self.rank + " of " + self.suit
```

```python
### Creating a deck class
class Deck():

    def __init__(self):        
        self.all_cards = []

        for suit in suits:
            for rank in ranks:
                created_card = Card(suit,rank)
                self.all_cards.append(created_card)

    def shuffle(self):
        random.shuffle(self.all_cards)

    def deal(self):
        return self.all_cards.pop()
        
```

```python
class Player():
    
    def __init__(self,name):
        self.cards = []
        self.name = name
        
    def add_card(self,card):
        if type(card) == type([]):
            self.cards.extend(card)
        else:
            self.cards.append(card)

    def remove_card(self):
        return self.cards.pop(0)

    def __str__(self):
        return f'Player {self.name} has {len(self.cards)} cards'
```

```python
def game():
    ### Getting Player Name 
    name1 = input('Input Player 1 name')
    clear_output()
    name2 = input('Input Player 2 name')
    clear_output()
    ### Creating Players
    p1 = Player(name1)
    p2 = Player(name2)
    
    ### Creating Deck
    new_deck = Deck()
    new_deck.shuffle()

    ### Splitting cards between the players 
    for i in range(26):
        p1.add_card(new_deck.deal())
        p2.add_card(new_deck.deal())
    game_on = True
    ### War
    round_num = 0
    while game_on:
        round_num += 1
        print(f'Round {round_num}')

        if len(p1.cards) == 0:
            print(f'Player {name1} is out of cards ! {name2} wins ! ')
            game_on = False
        elif len(p2.cards) == 0:
            print(f'Player {name2} is out of cards ! {name1} wins ! ')
            game_on = False

        ## Start new round
        player_one_cards = []
        player_one_cards.append(p1.remove_card())

        player_two_cards = []
        player_two_cards.append(p2.remove_card())

        #while at war
        at_war = True
        while at_war:
            if player_one_cards[-1].value > player_two_cards[-1].value:
                print(f'Player {name1} wins the round ! ')
                p1.add_card(player_one_cards)
                p1.add_card(player_two_cards)
                at_war = False
            
            elif player_one_cards[-1].value < player_two_cards[-1].value:
                print(f'Player {name2} wins the round ! ')
                p2.add_card(player_one_cards)
                p2.add_card(player_two_cards)
                at_war = False

            else:
                print('WAR!!')
                if len(p1.cards) < 5:
                    print('Player 1 doesnt have enough cards , player 2 wins')
                    game_on = False
                    break

                elif len(p2.cards) < 5:
                    print('Player 2 doesnt have enough cards , player 1 wins')
                    game_on = False
                    break
                else:
                    print('Both players got same value cards ! Draw 5 each !')
                    for i in range(5):
                        player_one_cards.append(p1.remove_card())
                        player_two_cards.append(p2.remove_card())      
```
