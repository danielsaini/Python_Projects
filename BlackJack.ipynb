```python
suits = ('Heart','Diamond','Club','Spade')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Joker','Queen','King','Ace')
values = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Joker':10,'Queen':10,'King':10,'Ace':11}
playing = True
```

```python
class Card():

    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]

    def __str__(self):
        return f'{self.rank} of {self.suit}'
```

```python
class Deck():
    
    def __init__(self):
        self.new_deck = []
        for suit in suits:
            for rank in ranks:
                self.new_deck.append(Card(suit,rank))

    def __str__(self):
        return f'Deck has {len(self.new_deck)} cards'

    def shuffle(self):
         random.shuffle(self.new_deck)

    def deal(self):
        single_card = self.new_deck.pop()
        return single_card
```

```python
class Hand:
    def __init__(self):
        self.cards = []
        self.value = 0
        self.aces = 0

    def add_card(self,card):
        #from deck.deal() -----> single card( suit , rank ) 
        self.cards.append(card)
        self.value += values[card.rank]

        ## Track aces
        if card.rank == 'Ace':
            self.aces += 1

    def abjust_for_ace(self):
        
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
class Chips:

    def __init__(self,total=100):
        self.total = total
        self.bet = 0

    def win_bet(self):
        self.total += self.bet

    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips):
    while True:
        try:
            chips.bet = int(input('How many chips would you like to bet?'))
        except:
            print('Sorry please provide an integer')
        else:
            if chips.bet > chips.total:
                print(f'Sorry , you dont have enoug chips ! you have {chips.total}')
            else:
                break
```

```python
def hit(deck,hand):

    hand.add_card(deck.deal())
    hand.abjust_for_ace()
```

```python
def hit_or_stand(deck,hand):
    global playing

    while True:
        x = input('Hit or Stand? Enter h or s ')

        if x[0].lower() == 'h':

            hit(deck,hand)

        elif x[0].lower() == 's':
            print("Player Stands , Dealerts Turn")
            playing = False

        else:
            print('Sorry , i did not understand that , please enter h or s only!')
            continue

        break
```
