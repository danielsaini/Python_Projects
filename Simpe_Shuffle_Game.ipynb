```python
from random import shuffle
```

```python
my_list = ['','O','']
```

```python
def ball_shuffle(my_list):
    shuffle(my_list)
    return my_list
```

```python
def input_check():
    a = ''
    while a not in ['0','1','2']:
        a = input('Please choose a position : 0 , 1 or 2')

    return int(a)
```

```python
def final_compare(mixed_list,guess):
    if mixed_list[guess] == 'O':
        print('Correct!')
    else:
        print('Incorrect!')
```

```python
### Mixing the list
mixed_list = ball_shuffle(my_list)
### Getting users guess
guess = input_check()
### Comparing 
final_compare(mixed_list,guess)
```
